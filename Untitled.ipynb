```python
# Perform data analysis
# Example: Calculate total allocated and distributed quantities of rice and wheat
total_allocated_rice = may_data['Rice Allocated'].sum()
total_distributed_rice = may_data['Rice Distributed'].sum()

total_allocated_wheat = may_data['Wheat Allocated'].sum()
total_distributed_wheat = may_data['Wheat Distributed'].sum()

# Print the results
print("Total Allocated Rice [Kg]:", total_allocated_rice )

print("Total Distributed Rice [Kg]:", total_distributed_rice)
print("Total Allocated Wheat [Kg]:", total_allocated_wheat)
print("Total Distributed Wheat [Kg]:", total_distributed_wheat)
```

```python
# Select the column for which you want to create a bar chart
column_name = 'Rice Allocated'

# Plot the bar chart
plt.figure(figsize=(20,6))
plt.bar(may_data.index, may_data[column_name])
plt.xlabel('Dealer')
plt.ylabel('Rice Allocated')
plt.title('Bar Chart of Rice Allocated')

# Rotate x-axis labels if needed
plt.xticks(rotation='vertical')

# Display the bar chart
plt.show()
```

```python
# Select the columns for which you want to create a bar chart
columns = ['Rice Allocated', 'Rice Distributed', 'Wheat Allocated', 'Wheat Distributed']

# Create subplots for each column
fig, axes = plt.subplots(nrows=len(columns), ncols=1, figsize=(8, 6))
fig.tight_layout()

# Iterate over the columns and plot the bar chart
for i, column in enumerate(columns):
    ax = axes[i]
    ax.bar(may_data.index, may_data[column])
    ax.set_xlabel('Dealer')
    ax.set_ylabel(column)
    ax.set_title(f'Bar Chart of {column}')
    ax.tick_params(axis='x', rotation=90)

# Display the bar charts
plt.show()
```

```python
# Select the columns for which you want to create a bar chart
columns = ['Rice Allocated', 'Rice Distributed', 'Wheat Allocated', 'Wheat Distributed']

# Set the dealer names as the x-axis labels
dealers = may_data['Dealer Name']

# Create subplots for each column
fig, axes = plt.subplots(nrows=len(columns), ncols=1, figsize=(20, 10))
fig.tight_layout()


# Adjust the gap between subplots
# plt.subplots_adjust(hspace=2.5)  # Increase or decrease the value as needed



# Iterate over the columns and plot the bar chart for each dealer
for i, column in enumerate(columns):
    ax = axes[i]

    ax.bar(dealers, may_data[column])
    ax.set_xlabel('Dealer')
    ax.set_ylabel(column)
    ax.set_title(f'Bar Chart of {column}')
    ax.tick_params(axis='x', rotation=90)

# Display the bar charts
plt.show()
```

```python
# Calculate the total allocated quantity for each dealer
may_data['Total Allocated'] = may_data['Rice Allocated'] + may_data['Wheat Allocated']

# Calculate the total distributed quantity for each dealer
may_data['Total Distributed'] = may_data['Rice Distributed'] + may_data['Wheat Distributed']

# Sort the dealers based on total allocations in descending order
top_allocated_dealers = may_data.sort_values('Total Allocated', ascending=False)

# Sort the dealers based on total distributions in descending order
top_distributed_dealers = may_data.sort_values('Total Distributed', ascending=False)

# Plot the top allocated dealers
plt.bar(top_allocated_dealers['Dealer Name'][:10], top_allocated_dealers['Total Allocated'][:10])
plt.xlabel('Dealer')
plt.ylabel('Total Allocated')
plt.title('Top 10 Dealers by Total Allocations')
plt.xticks(rotation='vertical')
plt.show()

# Plot the top distributed dealers
plt.bar(top_distributed_dealers['Dealer Name'][:10], top_distributed_dealers['Total Distributed'][:10])
plt.xlabel('Dealer')
plt.ylabel('Total Distributed')
plt.title('Top 10 Dealers by Total Distributions')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Calculate the discrepancies for Rice and Wheat
may_data['Rice Discrepancy'] = may_data['Rice Allocated'] - may_data['Rice Distributed']
may_data['Wheat Discrepancy'] = may_data['Wheat Allocated'] - may_data['Wheat Distributed']

# Calculate the total discrepancy for each dealer
may_data['Total Discrepancy'] = may_data['Rice Discrepancy'] + may_data['Wheat Discrepancy']

# Sort the dealers based on total discrepancy in descending order
sorted_data = may_data.sort_values('Total Discrepancy', ascending=False)

# Plot the top 10 dealers with the highest discrepancy
plt.bar(sorted_data['Dealer Name'][:10], sorted_data['Total Discrepancy'][:10])
plt.xlabel('Dealer')
plt.ylabel('Total Discrepancy')
plt.title('Top 10 Dealers with Highest Discrepancy')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Transaction Counts
transaction_counts = may_data.groupby('Dealer Name')['Transaction Count (By Invoice)'].count().reset_index()
transaction_counts = transaction_counts.sort_values('Transaction Count (By Invoice)', ascending=False)

# Transaction Values
transaction_values = may_data.groupby('Dealer Name')['Transaction Count (By Invoice)'].sum().reset_index()
transaction_values = transaction_values.sort_values('Transaction Count (By Invoice)', ascending=False)

# Transaction Distributions
rice_distributions = may_data.groupby('Dealer Name')['Rice Distributed'].sum().reset_index()
rice_distributions = rice_distributions.sort_values('Rice Distributed', ascending=False)

wheat_distributions = may_data.groupby('Dealer Name')['Wheat Distributed'].sum().reset_index()
wheat_distributions = wheat_distributions.sort_values('Wheat Distributed', ascending=False)

# Plotting Transaction Counts
plt.bar(transaction_counts['Dealer Name'][:10], transaction_counts['Transaction Count (By Invoice)'][:10])
plt.xlabel('Dealer')
plt.ylabel('Transaction Count')
plt.title('Top 10 Dealers by Transaction Count')
plt.xticks(rotation='vertical')
plt.show()

# Plotting Transaction Values
plt.bar(transaction_values['Dealer Name'][:20], transaction_values['Transaction Count (By Invoice)'][:20])
plt.xlabel('Dealer')
plt.ylabel('Transaction Value')
plt.title('Top 20 Dealers by Transaction Value')
plt.xticks(rotation='vertical')
plt.show()

# Plotting Rice Distributions
plt.bar(rice_distributions['Dealer Name'][:20], rice_distributions['Rice Distributed'][:20])
plt.xlabel('Dealer')
plt.ylabel('Rice Distributed')
plt.title('Top 20 Dealers by Rice Distribution')
plt.xticks(rotation='vertical')
plt.show()

# Plotting Wheat Distributions
plt.bar(wheat_distributions['Dealer Name'][:20], wheat_distributions['Wheat Distributed'][:20])
plt.xlabel('Dealer')
plt.ylabel('Wheat Distributed')
plt.title('Top 20 Dealers by Wheat Distribution')
plt.xticks(rotation='vertical')
plt.show()
```
